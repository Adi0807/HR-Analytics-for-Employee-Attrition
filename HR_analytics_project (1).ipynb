```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('Cleaned_EmployeeAttrition.csv')
```

```python
data.head()
```

```python
data.info()
```

```python
data.isnull().sum()
```

# Attrition Rate by Department

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Department', y='Attrition', data=data, ci=None)
plt.title('Attrition Rate by Department')
plt.xlabel('Department')
plt.ylabel('Attrition Rate')
plt.show()
```

# Attrition Rate by Age Group

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='AgeGroup', y='Attrition', data=data, ci=None)
plt.title('Attrition Rate by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Attrition Rate')
plt.show()
```

# Attrition Rate by Job Role

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='JobRole', y='Attrition', data=data, ci=None)
plt.xticks(rotation=45)
plt.title('Attrition Rate by Job Role')
plt.xlabel('Job Role')
plt.ylabel('Attrition Rate')
plt.show()
```

# Monthly Income Distribution by Attrition

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Attrition', y='MonthlyIncome', data=data)
plt.title('Monthly Income Distribution by Attrition Status')
plt.xlabel('Attrition (1=Yes, 0=No)')
plt.ylabel('Monthly Income')
plt.show()
```
